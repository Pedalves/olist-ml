## Products per order

```python
dt_orders.head()
```

```python
dt_orders_items = pd.read_csv('../dataset/olist_order_items_dataset.csv', index_col=0)
dt_orders_items.head()
```

```python
dt_products = pd.read_csv('../dataset/olist_products_dataset.csv', index_col=0)
dt_products.head()
```

```python
order_products_info = {}

for c in dt_products.columns:
    order_products_info[c] = dt_orders_items.product_id.map(dt_products[c])
    order_products_info[c].name = dt_products[c].name
```

```python
orders_products = pd.DataFrame(order_products_info)
orders_products.head()
```

## Customers Info

```python
dt_payments = pd.read_csv('../dataset/olist_order_payments_dataset.csv', index_col=0)
```

```python
dt_review_costumer = pd.concat([order_unique_costumer, review_by_order], axis=1)
dt_review_costumer.index.name = 'order_id'
dt_review_costumer.head()
```

```python
order_payment_product = orders_products.reset_index().merge(
                        dt_payments.reset_index()).merge(
                        dt_orders_items[["price","freight_value"]].reset_index())
```

```python
order_payment_product.head()
```

```python
dt_all_info = order_payment_product.merge(dt_review_costumer.reset_index())
dt_all_info.head()
```

```python
mean_items_per_order = dt_all_info.groupby(['customer_unique_id','order_id'])['review_score'].count().reset_index().groupby('customer_unique_id')['review_score'].mean()
number_of_orders = dt_all_info.groupby('customer_unique_id')['review_score'].count()
```

```python
dt_all_info.drop(['product_category_name', 'order_id', 'payment_type'], axis=1, inplace = True)
dt_all_info.head()
```

```python
customer_features = dt_all_info.groupby('customer_unique_id').mean()
customer_features['mean_items_per_order'] = mean_items_per_order
customer_features['number_of_orders'] = number_of_orders
```

```python
customer_features.head()
```

## Clustering

```python
customer_features['size_cm3'] = customer_features.product_height_cm*customer_features.product_width_cm*customer_features.product_length_cm
customer_features.dropna(inplace=True)
customer_features_original = customer_features.copy()

#customer_features.drop(['freight_value','payment_sequential','product_name_lenght', 'product_description_lenght', ], axis = 1, inplace=True)

        
customer_features.drop(['product_height_cm','product_width_cm',
                           'product_length_cm'], axis = 1, inplace=True)

customer_features.head()
```

```python
#x_features = customer_features.values
x_features = customer_features.drop('review_score',axis=1).values
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_samples, silhouette_score

scaler = MinMaxScaler()
clusterer = KMeans(n_clusters=15, random_state=42)
```

```python
scaled_features = scaler.fit_transform(x_features)
```

```python
cluster_labels = clusterer.fit_predict(scaled_features)
```

```python
customer_features['clusters'] = cluster_labels
customer_features.head()
```

```python
fig = px.histogram(customer_features, x="review_score", color="clusters", opacity=0.3, marginal="violin")
fig.show()
```
