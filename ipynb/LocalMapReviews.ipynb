## Dataset geolocation

```python
dt_geo = pd.read_csv('../dataset/olist_geolocation_dataset.csv')
dt_geo.head()
```

```python
dt_geo_unique = dt_geo.groupby('geolocation_zip_code_prefix').head(1)
dt_geo_unique = dt_geo_unique.set_index('geolocation_zip_code_prefix')
dt_geo_unique.head()
```

```python
city_state = (dt_geo_unique.geolocation_city +"-"+ dt_geo_unique.geolocation_state)
city_state.head()
```

## Customers dataset

```python
dt_customers = pd.read_csv('../dataset/olist_customers_dataset.csv', index_col=0)
dt_customers.head()
```

## Review mean by location

```python
dt_reviews = pd.read_csv('../dataset/olist_order_reviews_dataset.csv', index_col=0)
```

```python
dt_reviews.head()
```

```python
dt_reviews.shape
```

```python
order_review = dt_reviews.groupby('order_id')['review_score'].mean()
```

```python
dt_orders = pd.read_csv('../dataset/olist_orders_dataset.csv', index_col = 0)
```

```python
dt_orders.head()
```

```python
customer_review = pd.concat([dt_orders.customer_id, order_review], axis=1).groupby('customer_id').mean()
```

```python
zip_review = pd.concat([dt_customers.customer_zip_code_prefix, 
                        customer_review], axis=1).groupby('customer_zip_code_prefix').mean()
```

## Orders Count by Location

```python
df_orders_location = pd.concat([dt_customers.customer_zip_code_prefix.value_counts(),
           dt_geo_unique.geolocation_lat, 
           dt_geo_unique.geolocation_lng,
                               zip_review,
                               city_state],axis=1).dropna()
```

```python
df_orders_location.columns = ['quantity', 'geolocation_lat', 'geolocation_lng', 'review_score_mean','city']
df_orders_location['quantity_log'] = np.log(df_orders_location['quantity'])
df_orders_location.head()
```

## Dash showing quantity of orders and mean review

```python
import plotly.express as px
px.set_mapbox_access_token(open("../.mapbox_token").read())

fig = px.scatter_mapbox(df_orders_location, hover_name = 'city',lat="geolocation_lat", lon="geolocation_lng", size="quantity_log",
                  color_continuous_scale=px.colors.diverging.RdYlBu, size_max=5, zoom=2.5, color="review_score_mean")
fig.show()
```

## Dash showing quantity of orders and mean review < 3

```python
df_orders_location_bad = df_orders_location.loc[df_orders_location['review_score_mean']<3]
df_orders_location_bad.head()
```

```python
fig = px.scatter_mapbox(df_orders_location_bad, hover_name = 'city',lat="geolocation_lat", lon="geolocation_lng", size="quantity_log",
                  color_continuous_scale=px.colors.sequential.solar, size_max=5, zoom=2.5, color="review_score_mean", mapbox_style = 'carto-darkmatter')
fig.show()
```
